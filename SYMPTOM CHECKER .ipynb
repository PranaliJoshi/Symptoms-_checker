```python
from sklearn.preprocessing import LabelEncoder # used label encoder to simply convert each value in a column to a number.
from tqdm import tqdm

# used coo_matrix to convert spare matrix into COOrdinate format.
# used csr_matrix to Compressed Sparse Column matrix

from scipy.sparse import coo_matrix, csr_matrix  

def read_data(filename):
    # usecols load specific columns into dataframe 
    # read in triples of user/song/play from the input dataset
     #used skiprows Line numbers to skip while reading csv
    data = pd.read_csv(filename,usecols=[0,1,2],names=['user', 'song','plays'],skiprows=1)
    data=data.dropna(axis=0, how='any')  #drop nan values
    data['plays']=data['plays']+1  
    print(data.head())
    
    # maped each diagnose and symptoms to a unique numeric value
    data['user'] = data['user'].astype("category")
    data['song'] = data['song'].astype("category")
    
    # create a sparse matrix of all the symptoms/diagnose
    plays = coo_matrix((data['plays'].astype(float),
                        (data['song'].cat.codes.copy(),
                        data['user'].cat.codes.copy())))
    
    return data, plays,data.groupby(['song']).plays.sum(),data['user'].cat.codes.copy()

data,matrix,songsd,user = read_data("C:/Users/Lenovo/Desktop/Disease_detection_through_symptoms/diffsydiw.csv")
data.head(10)
# user = symptoms 
# song = diagnose
```
